```python

# --- 3. Serial & Logic ---
PORT = '/dev/cu.usbserial-0001'
GLOBAL_BREAK = 0.4
last_global_trigger = 0

THRESHOLDS_1 = [100, 100, 100, 100, 100, 100, 100]

try:
    ser.close()
except:
    pass

ser = serial.Serial(PORT, 115200, timeout=0.01)
print("Organic 7-Sensor Synth Online! Velocity-sensitive volume enabled.")



try:
    while True:
        line = ser.readline().decode('utf-8', errors='ignore').strip()

        if line and ',' in line:
            try:
                vals = [int(v) for v in line.split(',')]
                now = time.time()

                if (now - last_global_trigger) < GLOBAL_BREAK:
                    continue

                for i in range(len(vals)):

                    if vals[i] > THRESHOLDS_1[i]:

                        # # Map sensor value 100–300 -> volume 0.2–1.0
                        # strength = max(100, min(vals[i], 1000))
                        # volume = 0.1 + ((strength - 100) / 900) * 0.9

                        # notes[i].set_volume(volume)
                        # notes[i].play()
                        if i == 0:
                            client.send_message("/ind/snare", 1)
                        elif i == 1:
                            client.send_message("/ind/kick", 1)
                        elif i == 2:
                            client.send_message("/altj/tabla", 1)
                        elif i == 3:
                            client.send_message("/altj/pluck", 1)
                        elif i == 4:
                            client.send_message("/ind/epiano", 1)
                        elif i == 5:
                            client.send_message("/ind/jangle", 1)
                        elif i == 6:
                            client.send_message("/altj/tanpura", 1)
                        elif i == 7:
                            client.send_message("/sparkle", 1)

                        print(
                            f"PIEZO {i} triggered! "
                            f"Signal: {vals[i]}, "
                        )

                        last_global_trigger = now
                        break

            except (ValueError, IndexError):
                continue

finally:
    ser.close()
    pygame.mixer.quit()
```
